```python
import pandas as pd
import numpy as np

df_evasao = pd.read_csv('./dados/dataset.csv')
df_evasao.head()
```

```python
df_formacao_pais = pd.read_csv('./dados/escolariedade.csv', sep = ";")
df_formacao_pais.head()
```

```python
df_formacao_pais.shape
```

```python
df_evasao.shape
```

```python
df_merge = pd.concat([df_evasao, df_formacao_pais], axis=1).copy()
df_merge.shape
```

```python
def formatar(valor):
    valor = valor*100
    return "{:.2f}".format(valor)
```

```python
df_merge["Evasão"] = df_merge["Target"].map({"Dropout":1,"Graduate":0, "Enrolled":0})
comparacao = df_merge.pivot_table(index="escolaridade_mae",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Escolaridade mãe', 'Evasão (%)', 'Evasão absoluta', 'Total alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="escolaridade_pai",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Escolaridade pai', 'Evasão (%)', 'Evasão absoluta', 'Total alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="ocupacao_mae",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Ocupação da mãe', 'Evasão (%)', 'Evasão absoluta', 'Total de alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="ocupacao_pai",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Ocupação do pai', 'Evasão (%)', 'Evasão absoluta', 'Total de alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
df = pd.read_csv('./dados/dataset.csv')
df = df[["Debtor", "Tuition fees up to date", "Target"]]
df["Evadiu"] = df["Target"].map({"Dropout":"Sim","Graduate":"Não","Enrolled":"Não"})
df.loc[df["Debtor"] == 0, "Debtor"] = "Não"
df.loc[df["Debtor"] == 1, "Debtor"] = "Sim"
df.loc[df["Tuition fees up to date"] == 0, "Tuition fees up to date"] = "Não"
df.loc[df["Tuition fees up to date"] == 1, "Tuition fees up to date"] = "Sim"
df = df.rename(columns={"Debtor":"Devedor", "Tuition fees up to date":"Mensalidades em dia"})
df
```

```python
debtor = df.loc[(df.Devedor == "Sim") & (df["Mensalidades em dia"] == "Sim")]
debtor_and_pendent = df.loc[(df.Devedor == "Sim") & (df["Mensalidades em dia"] == "Não")]
up_to_date = df.loc[(df.Devedor == "Não") & (df["Mensalidades em dia"] == "Sim")]
pendent = df.loc[(df.Devedor == "Não") & (df["Mensalidades em dia"] == "Não")]
a = debtor.groupby(["Evadiu"]).count()
b = debtor_and_pendent.groupby(["Evadiu"]).count()
c = up_to_date.groupby(["Evadiu"]).count()
d = pendent.groupby(["Evadiu"]).count()
```
