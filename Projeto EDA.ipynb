```python
import pandas as pd
import numpy as np

df_evasao = pd.read_csv('./dados/dataset.csv')
df_evasao.head()
```

```python
df_formacao_pais = pd.read_csv('./dados/escolariedade.csv', sep = ";")
df_formacao_pais.head()
```

```python
df_formacao_pais.shape
```

```python
df_evasao.shape
```

```python
df_merge = pd.concat([df_evasao, df_formacao_pais], axis=1).copy()
df_merge.shape
```

```python
def formatar(valor):
    valor = valor*100
    return "{:.2f}".format(valor)
```

```python
df_merge["Evasão"] = df_merge["Target"].map({"Dropout":1,"Graduate":0})
comparacao = df_merge.pivot_table(index="escolaridade_mae",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Escolaridade mãe', 'Evasão (%)', 'Evasão absoluta', 'Total alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="escolaridade_pai",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Escolaridade pai', 'Evasão (%)', 'Evasão absoluta', 'Total alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="ocupacao_mae",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Ocupação da mãe', 'Evasão (%)', 'Evasão absoluta', 'Total de alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="ocupacao_pai",values=["Evasão"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Ocupação do pai', 'Evasão (%)', 'Evasão absoluta', 'Total de alunos']
comparacao['Evasão (%)'] = comparacao['Evasão (%)'].apply(formatar)
comparacao
```
