```python
import pandas as pd
import numpy as np

df_evasao = pd.read_csv('./dados/dataset.csv')
df_evasao.head()
```

```python
df_formacao_pais = pd.read_csv('./dados/escolariedade.csv', sep = ";")
df_formacao_pais.head()
```

```python
df_formacao_pais.shape
```

```python
df_evasao.shape
```

```python
df_merge = pd.concat([df_evasao, df_formacao_pais], axis=1).copy()
df_merge.shape
```

```python
def formatar(valor):
    valor = valor*100
    return "{:.2f}".format(valor)
```

```python
df_merge["EvasÃ£o"] = df_merge["Target"].map({"Dropout":1,"Graduate":0})
comparacao = df_merge.pivot_table(index="escolaridade_mae",values=["EvasÃ£o"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Escolaridade mÃ£e', 'EvasÃ£o (%)', 'EvasÃ£o absoluta', 'Total alunos']
comparacao['EvasÃ£o (%)'] = comparacao['EvasÃ£o (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="escolaridade_pai",values=["EvasÃ£o"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['Escolaridade pai', 'EvasÃ£o (%)', 'EvasÃ£o absoluta', 'Total alunos']
comparacao['EvasÃ£o (%)'] = comparacao['EvasÃ£o (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="ocupacao_mae",values=["EvasÃ£o"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['OcupaÃ§Ã£o da mÃ£e', 'EvasÃ£o (%)', 'EvasÃ£o absoluta', 'Total de alunos']
comparacao['EvasÃ£o (%)'] = comparacao['EvasÃ£o (%)'].apply(formatar)
comparacao
```

```python
comparacao = df_merge.pivot_table(index="ocupacao_pai",values=["EvasÃ£o"],aggfunc=["mean","sum","count"]).reset_index()
comparacao.columns = ['OcupaÃ§Ã£o do pai', 'EvasÃ£o (%)', 'EvasÃ£o absoluta', 'Total de alunos']
comparacao['EvasÃ£o (%)'] = comparacao['EvasÃ£o (%)'].apply(formatar)
comparacao
```
